```python
from bs4 import BeautifulSoup
import requests

req = requests.get('https://www.medrxiv.org/content/10.1101/2021.01.02.20248998v1.full')
soup = BeautifulSoup(req.content, 'html.parser')
meta = soup.find_all("meta", attrs = {"name" : "citation_reference"})
```

```python
 def parseAmpersandCitation(c):
    tokens = c.split("&")
    tokens = tokens[1].split(".")
    # author middle initial messes up split on ., sometimes need to go one over
    if(len(tokens[1]) < 5):
        return tokens[2]
    else:
        return tokens[1]

def parseEtAlCitation(c):
    tokens = c.split("et al.")
    return tokens[1].split(".")[0]

def parseSingleAuthorCitation(c):
    c = c.replace("?", ".")
    tokens = c.split(".")
    # author middle initial messes up split on ., sometimes need to go one over
    if(len(tokens[1]) < 5):
        return tokens[2]
    else:
        return tokens[1]
```

```python
def removeAuthors(s):
        
    def ampersand(s):
        tokens = s.split("&")
        return tokens[1].split(",", 1)[1].split(" ", 2)[-1]
    
    def etAl(s):
        tokens = s.split("et al.")[1]
        return tokens
    
    def other(s):
        tokens =  s.split(" ", 1)
        if "," in tokens[0]:
            return tokens[1].split(" ", 1)[1]
        else:
            return tokens[1]

    if "&" in s:
        return ampersand(s)
    elif "et al." in s:
        return etAl(s)
    else:
        return other(s)

def removeJournal(s):
    s = s.replace("?", ".")
    return s.split(".", 1)[0].strip()

for m in meta:  
    test = m['content']
    if "&" in test:
        print(test, "\n>")
        new = removeAuthors(test)
        print(new, "\n>")
        print(removeJournal(new), "\n\n\n")
    elif "et al." in test:
        print(test, "\n>")
        new = removeAuthors(test)
        print(new, "\n>")
        print(removeJournal(new), "\n\n\n")
    else:
        print(test, "\n>")
        new = removeAuthors(test)
        print(new, "\n>")
        print(removeJournal(new), "\n\n\n")
        print("\n\n\n")
        
```
